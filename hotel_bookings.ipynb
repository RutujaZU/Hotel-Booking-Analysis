# Data Analysis and Visualisation

```python
cancelled_perc =data['is_canceled'].value_counts(normalize=True)
cancelled_perc
```

```python
cancelled_perc =data['is_canceled'].value_counts(normalize=True)
print(cancelled_perc)

plt.figure(figsize=(5,4))
plt.title('reservation status count')
plt.bar(['Not canceled','Canceled'],data['is_canceled'].value_counts(),edgecolor='k',width=0.7)
plt.show()
```

```python
plt.figure(figsize=(8, 4))

# Create a count plot using seaborn
ax1 = sns.countplot(x='hotel', hue='is_canceled', data=data, palette='Blues')

# Get legend labels and handle
legend_labels, _ = ax1.get_legend_handles_labels()

# Place the legend outside the plot
ax1.legend(bbox_to_anchor=(1, 1))

# Set title and axis labels
plt.title('Reservation status in different hotels', size=10)
plt.xlabel('hotel')
plt.ylabel('number of reservations')

# Display the plot
plt.show()
```

```python
resort_hotel=data[data['hotel']== 'Resort Hotel']
resort_hotel['is_canceled'].value_counts(normalize=True)
```

```python
city_hotel=data[data['hotel']== 'City Hotel']
city_hotel['is_canceled'].value_counts(normalize=True)
```

```python
resort_hotel= resort_hotel.groupby('reservation_status_date')[['adr']].mean()
city_hotel= city_hotel.groupby('reservation_status_date')[['adr']].mean()
```

```python
plt.figure(figsize=(20, 8))
plt.title('Average Daily Rate in Resort Hotel & City Hotel', fontsize=30)

# Plotting the ADR for Resort Hotel
plt.plot(resort_hotel.index, resort_hotel['adr'], label='Resort Hotel')

# Plotting the ADR for City Hotel
plt.plot(city_hotel.index, city_hotel['adr'], label='City Hotel')

# Adding legend
plt.legend()

# Adding axis labels
plt.xlabel('Reservation Status Date')
plt.ylabel('Average Daily Rate')

# Show the plot
plt.show()
```

```python
# Extracting the month from the 'reservation_status_date' column
data['month'] = data['reservation_status_date'].dt.month

# Setting the size of the figure (plot)
plt.figure(figsize=(16, 8))

# Creating a count plot using Seaborn
ax1 = sns.countplot(x='month', hue='is_canceled', data=data, palette='bright')

# Retrieving legend handles and labels
legend_labels, _ = ax1.get_legend_handles_labels()

# Placing the legend outside the plot
ax1.legend(bbox_to_anchor=(1, 1))

# Setting the title of the plot
plt.title('Reservation status per month', size=20)

# Adding labels to the x-axis and y-axis
plt.xlabel('Month')
plt.ylabel('Number of reservations')

# Adding a legend with custom labels
plt.legend(['Not Canceled', 'Canceled'])

# Displaying the plot
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.title('ADR per month', fontsize=20)

# Filtering data for canceled reservations and grouping by month
sns.barplot(x='month', y='adr', data=data[data['is_canceled'] == 1].groupby('month')[['adr']].sum().reset_index())

plt.show()
```

```python
# Filtering data for canceled reservations
canceled_data = data[data['is_canceled'] == 1]

# Counting the number of canceled reservations for each country and selecting the top 10
top_10_country = canceled_data['country'].value_counts()[:10]

# Creating a pie chart
plt.figure(figsize=(8, 8))
plt.title('Top 10 countries with reservation canceled')

# Plotting the pie chart
plt.pie(top_10_country, autopct='%.2f', labels=top_10_country.index)

# Displaying the plot
plt.show()
```

```python
canceled_data_adr= canceled_data.groupby('reservation_status_date')[['adr']].mean()
canceled_data_adr.reset_index(inplace= True)
canceled_data_adr.sort_values('reservation_status_date' , inplace = True)

not_canceled_data=data[data['is_canceled'] ==0]
not_canceled_data_adr= not_canceled_data.groupby('reservation_status_date')[['adr']].mean()
not_canceled_data_adr.reset_index(inplace= True)
not_canceled_data_adr.sort_values('reservation_status_date' , inplace = True)

plt.figure(figsize =(20,6))
plt.title('Average Daily Rate')
plt.plot(not_canceled_data_adr['reservation_status_date'],not_canceled_data_adr['adr'],label = 'not canceled')
plt.plot(canceled_data_adr['reservation_status_date'],canceled_data_adr['adr'],label = 'canceled' )      
plt.legend()

plt.xlabel('Reservation Status Date')
plt.ylabel('Average Daily Rate')
```
